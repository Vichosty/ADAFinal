```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def mostrar_resultados():
    # Cargar datos
    try:
        df = pd.read_csv("resultados_mlst.csv")
    except FileNotFoundError:
        print("Error: No se encuentra 'resultados_mlst.csv'. Ejecuta la Parte 1 primero.")
        return

    # --- 1. TABLA DE CALIDAD (HOJAS) ---
    print("\n>>> TABLA DE CALIDAD (Hojas) <<<")
    tabla_calidad = df[["Caso", "Greedy_Hojas", "BL_Hojas", "ILS_Hojas"]].copy()
    tabla_calidad["Mejora ILS vs Greedy"] = tabla_calidad["ILS_Hojas"] - tabla_calidad["Greedy_Hojas"]
    # Calculamos el porcentaje de mejora
    tabla_calidad["% Mejora"] = ((tabla_calidad["Mejora ILS vs Greedy"] / tabla_calidad["Greedy_Hojas"]) * 100).round(1)
    print(tabla_calidad.to_markdown(index=False))

    # --- 2. TABLA DE TIEMPOS ---
    print("\n>>> TABLA DE TIEMPOS (Segundos) <<<")
    tabla_tiempo = df[["Caso", "Greedy_Tiempo", "BL_Tiempo", "ILS_Tiempo"]].copy()
    # Redondeamos para que se vea limpio
    tabla_tiempo = tabla_tiempo.round(4)
    print(tabla_tiempo.to_markdown(index=False))

    # --- 3. GRFICO DE BARRAS (Comparacin de Calidad) ---
    plt.figure(figsize=(12, 6))
    
    x = np.arange(len(df["Caso"]))
    width = 0.25
    
    plt.bar(x - width, df["Greedy_Hojas"], width, label='Greedy', color='#a0c4ff')
    plt.bar(x, df["BL_Hojas"], width, label='BL Simple', color='#ffcc99')
    plt.bar(x + width, df["ILS_Hojas"], width, label='ILS (Metaheurstica)', color='#90EE90')
    
    plt.xlabel('Casos de Prueba', fontsize=12)
    plt.ylabel('Nmero de Hojas (Mayor es mejor)', fontsize=12)
    plt.title('Comparacin de Algoritmos: Cantidad de Hojas', fontsize=14)
    plt.xticks(x, df["Caso"], rotation=15)
    plt.legend()
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    
    plt.tight_layout()
    plt.show()
    
    # --- 4. GRFICO DE TIEMPO (Comparacin de Costo) ---
    plt.figure(figsize=(12, 5))
    
    # Greedy (Azul)
    plt.plot(df["Caso"], df["Greedy_Tiempo"], marker='o', label='Greedy', color='blue', linewidth=2)
    
    # BL Simple (Naranja) - Aqu est el agregado!
    plt.plot(df["Caso"], df["BL_Tiempo"], marker='^', label='BL Simple', color='orange', linewidth=2, linestyle='--')
    
    # ILS (Verde)
    plt.plot(df["Caso"], df["ILS_Tiempo"], marker='s', label='ILS (Tiempo Fijo)', color='green', linewidth=2)
    
    plt.ylabel("Tiempo (segundos)", fontsize=12)
    plt.title("Costo Computacional: Greedy vs BL Simple vs ILS", fontsize=14)
    plt.xticks(rotation=15)
    plt.legend()
    plt.grid(True, linestyle=':', alpha=0.7)
    
    plt.tight_layout()
    plt.show()

# Ejecutar visualizador
mostrar_resultados()
```
